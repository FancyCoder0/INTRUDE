```python
# recall data prep
from git import *
from comp import *
from detect import *
from clf import *

# threshold = 0.60

repo_num = {}
repo_tot = {}

c = classify()
lastr = None

file = 'detection/result_on_topk_new_ok.txt'
fileout = 'detection/rresult_on_topk_new_ok_thres_z.txt'

with open(fileout, 'w') as f:
    pass

def check_ok_z(pull, pullA):
    if speed_up_check(pull, pullA):
        if abs((get_time(pullA["updated_at"]) - \
                get_time(pull["updated_at"])).days) >= 5 * 365: # more than 4 years
            return False
        if pull["user"]["id"] == pullA["user"]["id"]:
            return False
        if str(pull["number"]) in (get_pr_and_issue_numbers(pullA["title"]) + \
                                   get_pr_and_issue_numbers(pullA["body"])):
            return False
        if (pull["merged_at"] is not None) and (get_time(pull["merged_at"]) < get_time(pullA["created_at"])) and ((get_time(pullA["created_at"]) - get_time(pull["merged_at"])).days >= 14):
            return False
        
        return True
    
    return False
        
        
cnt = 0
with open(file) as f:
    for t in f.readlines():
        ps = t.replace(',','').replace('[','').replace(']','').split()            
        r, n1, n2 = ps[0], ps[1], ps[2]
        if n1 > n2:
            n1, n2 = n2, n1
        
        li = [int(x) for x in ps[3:]]

        if lastr != r:
            init_model_with_repo(r)
            lastr = r
            
        res = 0.0
        for x in li:
            x = str(x)
            if check_ok_z(get_pull(r, x), get_pull(r, n2)):
                res = c.predict_proba([get_sim(r, n2, x)])[0][1]
                break
        
        
        if res == 0:
            res = c.predict_proba([get_sim(r, n2, n1)])[0][1]
        
        
        cnt += 1
        if cnt % 100 == 0:
            print(cnt)
        with open(fileout,'a') as outf:
            print(r, n1, n2, res, file=outf)

print(cnt)
```
