```r
file = 'data/true_output.txt'

# pdf(file="data/true_boxplot.pdf")

library(ggpubr)

t <- read.table(file, header=TRUE)
p <- ggboxplot(t, x = "type", y = "probability", color = "type", palette = "jco",)
p + stat_compare_means()

# dev.off()
```
